```python
#@title Step 2: Dataset Loading, Custom Non-IID Partitioning, and Preprocessing (Add Client Count Print)

print("\n--- Step 2: Dataset Loading, Custom Non-IID Partitioning, and Preprocessing ---")

# --- Dataset Specifics for UCI HAR ---
NUM_FEATURES = 561
NUM_CLASSES = 6

# --- Function to load raw data (corrected for lowercase train/test folders) ---
def load_har_data_features(data_path, partition_type):
    """Loads features, labels, and subjects for train or test from X_*.txt."""
    features_path = os.path.join(data_path, partition_type, f'X_{partition_type}.txt')
    labels_path = os.path.join(data_path, partition_type, f'y_{partition_type}.txt')
    subjects_path = os.path.join(data_path, partition_type, f'subject_{partition_type}.txt')

    print(f"Loading {partition_type} data from {features_path}...")
    try:
        features = np.loadtxt(features_path)
        labels = np.loadtxt(labels_path, dtype=int) - 1
        subjects = np.loadtxt(subjects_path, dtype=int)
    except FileNotFoundError as e:
        print(f"Error loading file: {e}")
        raise

    print(f"  Loaded features shape: {features.shape}")
    assert features.shape[1] == NUM_FEATURES
    assert features.shape[0] == labels.shape[0] == subjects.shape[0]

    print(f"  Loaded {features.shape[0]} samples for {partition_type}.")
    return features, labels, subjects

# Load training data
train_features, train_labels, train_subjects = load_har_data_features(DATA_PATH, 'train')

# Get unique subject IDs - these will be our clients
all_subject_ids = sorted(list(np.unique(train_subjects)))
print(f"\nFound {len(all_subject_ids)} unique subjects (potential clients): {all_subject_ids}")

# --- Custom Non-IID Partitioning (Dirichlet) ---
def partition_data_by_subject_noniid(features, labels, subjects, subject_ids, alpha=0.5):
    """Partitions data by subject and applies non-IID distribution based on labels."""
    num_subjects = len(subject_ids)
    subject_label_proportions = np.random.dirichlet([alpha] * NUM_CLASSES, num_subjects)

    client_data_indices = {subject: [] for subject in subject_ids}
    label_indices = {label: np.where(labels == label)[0] for label in range(NUM_CLASSES)}
    label_counts = {label: len(indices) for label, indices in label_indices.items()}

    for label in range(NUM_CLASSES):
        if label_counts[label] == 0:
            continue
        current_label_indices = label_indices[label]
        subject_counts_for_label = (subject_label_proportions[:, label] * label_counts[label]).astype(int)
        subject_counts_for_label[0] += label_counts[label] - subject_counts_for_label.sum()
        subject_counts_for_label[subject_counts_for_label < 0] = 0

        np.random.shuffle(current_label_indices)

        start_idx = 0
        for i, subject in enumerate(subject_ids):
            count = subject_counts_for_label[i]
            count = min(count, len(current_label_indices) - start_idx)
            if count > 0:
                assigned_indices = current_label_indices[start_idx : start_idx + count]
                client_data_indices[subject].extend(assigned_indices.tolist())
                start_idx += count

    client_datasets = []
    client_ids_list = []
    client_data_sizes = {}

    print(f"\nPartitioning data using Dirichlet alpha={alpha}...")

    for subject in subject_ids:
        indices = client_data_indices[subject]
        if not indices:
             continue

        client_features = features[indices]
        client_labels = labels[indices]

        dataset = tf.data.Dataset.from_tensor_slices({
            'x': tf.constant(client_features, dtype=tf.float32),
            'y': tf.constant(client_labels, dtype=tf.int32)
        })

        client_datasets.append(dataset)
        client_ids_list.append(f'subject_{subject}')
        client_data_sizes[f'subject_{subject}'] = len(indices)

        if len(client_ids_list) <= 5 or subject == subject_ids[-1]:
             label_counts = collections.Counter(client_labels.tolist())
             print(f"  Client subject_{subject} data count: {len(indices)}, Label distribution: {dict(label_counts)}")

    print("\nCustom Non-IID partitioning complete.")
    print(f"Number of TFF client datasets created: {len(client_ids_list)}") # <--- ADDED THIS PRINT

    return client_datasets, client_ids_list, client_data_sizes, alpha


# --- Perform partitioning ---
simulated_client_subject_ids = all_subject_ids # Use all subjects as clients
# simulated_client_subject_ids = all_subject_ids[:10] # Or a subset (e.g., 10 clients)

train_client_datasets, train_client_ids, client_data_sizes, partition_alpha = partition_data_by_subject_noniid(
    train_features, train_labels, train_subjects, simulated_client_subject_ids, alpha=0.5 # You can change alpha here
)

print(f"\nCreated {len(train_client_ids)} TFF client datasets for training (Confirm count).")


# --- Define Preprocessing Function for UCI HAR Features ---
def preprocess_har_features(dataset):
    """Preprocesses UCI HAR feature vectors dataset for the model."""
    def batch_format_fn(element):
        """Format a batch of feature vectors and labels."""
        return collections.OrderedDict(
            x=element['x'],
            y=tf.one_hot(element['y'], NUM_CLASSES))

    return dataset.shuffle(buffer_size=100).batch(32).map(batch_format_fn).prefetch(tf.data.AUTOTUNE)

# --- Load and preprocess test data for evaluation ---
test_features, test_labels, test_subjects = load_har_data_features(DATA_PATH, 'test')

test_dataset = tf.data.Dataset.from_tensor_slices({
    'x': tf.constant(test_features, dtype=tf.float32),
    'y': tf.constant(test_labels, dtype=tf.int32)
})
test_dataset = preprocess_har_features(test_dataset).cache().prefetch(tf.data.AUTOTUNE)

print("\nTest dataset loaded and preprocessed using features.")


# --- Determine input_spec *after* data loading and preprocessing function is defined ---
if not train_client_datasets:
     print("Warning: No client datasets created. input_spec may not be determined correctly.")
     # Set a placeholder input_spec or raise error if strictly needed later
     # Let's set a default if possible, though this indicates a fundamental data issue
     input_spec = collections.OrderedDict([('x', tf.TensorSpec(shape=(None, NUM_FEATURES), dtype=tf.float32)),
                                           ('y', tf.TensorSpec(shape=(None, NUM_CLASSES), dtype=tf.float32))])
     print(f"Using default input_spec due to no client datasets: {input_spec}")
else:
    example_raw_dataset = train_client_datasets[0]
    example_preprocessed_dataset = preprocess_har_features(example_raw_dataset).take(1)
    input_spec = example_preprocessed_dataset.element_spec

    print("\nDataset loading, partitioning, and preprocessing complete.")
    print(f"Determined input_spec for model_fn: {input_spec}")
```

```python
#@title Step 3: Model Definition (Corrected for UCI HAR Feature Vectors - Using Pre-calculated input_spec)

print("\n--- Step 3: Model Definition (Corrected for UCI HAR Feature Vectors) ---")

# --- Model Definition for UCI HAR Feature Vectors ---
# Based on Step 2, input data to the model will be batches of feature vectors with shape (None, NUM_FEATURES).
# The Keras model itself should be defined to accept a single example's shape (NUM_FEATURES,).

def create_har_keras_model():
  """Creates a simple Keras model for UCI HAR feature vectors."""
  model = tf.keras.models.Sequential([
      # Input layer expecting feature vectors of size NUM_FEATURES (561)
      tf.keras.layers.InputLayer(input_shape=(NUM_FEATURES,)),
      tf.keras.layers.Dense(256, activation='relu'),
      tf.keras.layers.Dropout(0.5), # Added dropout for regularization
      tf.keras.layers.Dense(128, activation='relu'),
      tf.keras.layers.Dropout(0.5), # Added dropout
      tf.keras.layers.Dense(NUM_CLASSES, activation='softmax') # Output layer for classification (6 classes)
  ])
  return model

def model_fn():
  """Returns a Keras model and wraps it as a tff.learning.Model."""
  keras_model = create_har_keras_model()

  # Use the input_spec that was calculated *outside* this function in Step 2.
  # This avoids capturing EagerTensors during TFF's graph building.
  # The variable 'input_spec' must be available in the scope where model_fn is defined.
  if 'input_spec' not in globals() and 'input_spec' not in locals():
       raise NameError("Error: 'input_spec' variable not found. Please ensure Step 2 ran correctly.")

  print(f"Using pre-calculated input_spec in model_fn: {input_spec}")

  # Wrap the Keras model using TFF's helper function
  return tff.learning.models.from_keras_model(
      keras_model,
      input_spec=input_spec, # Use the correct input spec derived from preprocessed data in Step 2
      loss=tf.keras.losses.CategoricalCrossentropy(), # Loss function for multi-class classification with one-hot labels
      metrics=[tf.keras.metrics.CategoricalAccuracy()]) # Metric to track during training

print("HAR Keras model defined (Dense layers for feature vectors) and wrapped for TFF.")
```

```python
#@title Step 4: Federated Learning Process Setup (FedAvg - Explicit build_fed_eval)

print("\n--- Step 4: Federated Learning Process Setup (FedAvg) ---")

# --- Print Versions for Debugging ---
print(f"TensorFlow version: {tf.__version__}")
print(f"TensorFlow Federated version: {tff.__version__}")


# --- Federated Learning Process Setup ---
def client_optimizer_fn():
  """Returns the client Keras optimizer."""
  return tf.keras.optimizers.Adam(learning_rate=0.001)

def server_optimizer_fn():
  """Returns the server Keras optimizer."""
  return tf.keras.optimizers.SGD(learning_rate=1.0)


print(f"\nType of model_fn: {type(model_fn)}")
print(f"Is client_optimizer_fn callable? {callable(client_optimizer_fn)}")
print(f"Is server_optimizer_fn callable? {callable(server_optimizer_fn)}")
print("Attempting to build federated_algorithm...")

# Build the federated training process
federated_algorithm = None
try:
    # In TFF 0.80.0, client_epochs might be passed during the build,
    # or handled automatically by dataset structure if available.
    # Let's try building without explicitly telling it about per-client epochs here.
    federated_algorithm = tff.learning.algorithms.build_weighted_fed_avg(
        model_fn=model_fn,
        client_optimizer_fn=client_optimizer_fn,
        server_optimizer_fn=server_optimizer_fn
        # Remove client_weighting='num_examples' if it was there implicitly
        # The default for build_weighted_fed_avg is typically 'num_examples' anyway.
    )
    print("\nFederated training algorithm built successfully.")
    # Print type signatures if available (they should be for the training algo)
    print(f"Initial state type: {federated_algorithm.initialize.type_signature.result}")
    print(f"Next step function type: {federated_algorithm.next.type_signature}")

except Exception as e:
    print(f"\nERROR: Failed to build federated_algorithm.")
    print(f"Error Type: {type(e).__name__}")
    print(f"Error Message: {e}")
    # Let the exception propagate if building the main algorithm fails
    raise


# --- Try building the evaluation process ---
evaluation_process = None # Initialize to None
print("\nAttempting to build federated evaluation process...")
try:
    # Explicitly use build_fed_eval as that seemed to work partially before
    evaluation_process = tff.learning.algorithms.build_fed_eval(model_fn=model_fn)
    print("Federated evaluation process built using tff.learning.algorithms.build_fed_eval.")

    # CORRECTED: Do not try to access .type_signature directly on evaluation_process
    # print(f"Evaluation process type: {evaluation_process.type_signature}")

except Exception as e:
    print(f"ERROR: Failed to build federated evaluation process.")
    print(f"Error Type: {type(e).__name__}")
    print(f"Error Message: {e}")
    # Evaluation failed, but training algorithm *might* still be usable.
    evaluation_process = None # Set to None if eval build failed


print("\nStep 4 complete.")
```

```python
#@title Step 5: Simulation Configuration

print("\n--- Step 5: Simulation Configuration ---")

# --- Simulation Parameters ---
num_rounds = 50 # Total number of federated learning rounds
# Determine the number of clients participating per round.
# Ensure clients_per_round is not more than the total number of available clients.
num_total_clients = len(train_client_ids)
clients_per_round = min(10, num_total_clients) # Participate with max 10 clients per round, or fewer if less than 10 are available

print(f"Simulation Parameters:")
print(f"  Total clients available: {num_total_clients}")
print(f"  Number of rounds: {num_rounds}")
print(f"  Clients participating per round: {clients_per_round}")

# --- Client Heterogeneity Configuration ---
# Simulate variation in local work (local epochs)
min_local_epochs = 3
max_local_epochs = 7

local_epochs_per_client = {}
if num_total_clients > 0:
    local_epochs_per_client = {
        client_id: random.randint(min_local_epochs, max_local_epochs)
        for client_id in train_client_ids
    }

    # Identify and configure a straggler client
    # Ensure there's at least one client to pick from
    straggler_client_id = train_client_ids[random.randint(0, num_total_clients - 1)]
    straggler_local_epochs = max_local_epochs * 3 # Make straggler do significantly more work
    local_epochs_per_client[straggler_client_id] = straggler_local_epochs # Assign straggler epochs

    print(f"\nClient Heterogeneity Simulation:")
    print(f"  Local epochs assigned randomly between {min_local_epochs} and {max_local_epochs} per client.")
    print(f"  Straggler client ID: {straggler_client_id}")
    print(f"  Straggler local epochs: {straggler_local_epochs}")
else:
    straggler_client_id = None
    print("\nNot enough clients to configure straggler or varying epochs.")


# --- Partial Participation Simulation ---
# This is handled implicitly in the training loop (Step 6) by randomly sampling
# `clients_per_round` from `train_client_ids` each round.
print(f"\nPartial Participation Simulation:")
print(f"  Each round, {clients_per_round} clients will be randomly selected (handled in training loop).")

# --- Adversarial Client Simulation ---
# Select one client to be adversarial.
# Ensure there are enough unique clients to pick from after potentially selecting a straggler.
adversary_client_id = None
if num_total_clients > (1 if straggler_client_id is not None else 0): # Need at least one client not potentially the straggler
    available_clients_for_adversary = list(train_client_ids)
    if straggler_client_id is not None and straggler_client_id in available_clients_for_adversary:
        available_clients_for_adversary.remove(straggler_client_id)

    if available_clients_for_adversary: # Check if there are any clients left to pick from
         adversary_client_id = random.choice(available_clients_for_adversary)

    print(f"\nAdversarial Client Simulation:")
    print(f"  Adversary client ID: {adversary_client_id}")
    print(f"  Attack type: Simple Label Flipping (e.g., activity 0 -> 1, 1 -> 2, ..., 5 -> 0)")

    # Function to apply label flipping for the adversary client (adapted for UCI HAR features)
    # This function works on the batched, preprocessed dataset format from preprocess_har_features.
    def apply_har_label_flip(dataset):
        """Applies simple label flipping (e.g., 0->1, 1->2, ..., 5->0) to a HAR dataset."""
        def flip_label(element):
            # element['y'] is already one-hot encoded from preprocess_har_features
            original_label = tf.argmax(element['y'], axis=-1) # Get original class index from one-hot [batch_size,]
            flipped_label = (original_label + 1) % NUM_CLASSES # Flip to next class (cyclically) [batch_size,]
            return collections.OrderedDict(
                x=element['x'], # x is unchanged [batch_size, NUM_FEATURES]
                y=tf.one_hot(flipped_label, NUM_CLASSES) # Return flipped label as one-hot [batch_size, NUM_CLASSES]
            )
        # Apply the mapping to the dataset
        return dataset.map(flip_label)
else:
    print("\nNot enough clients to configure an adversary.")


# --- Secure Aggregation Simulation (Conceptual / Simple DP) ---
# TFF supports secure aggregation (`tff.federated_secure_sum`), but simulating the full cryptographic
# proof in this environment is not practical.
# We simulate the *effect* of a privacy mechanism like DP by conceptually adding noise
# to the aggregated updates or observing its impact on model utility.
# In this setup, we will conceptually state that aggregation happens securely/privately,
# and optionally add noise to the reported metrics or rely on the noise_std parameter
# if a future custom process uses it.
add_dp_noise = True # Set to True to conceptually add DP noise effect
dp_noise_std = 0.001 # Standard deviation of Gaussian noise - impacts simulated privacy/utility trade-off

print(f"\nSecure Aggregation Simulation:")
print(f"  Adding server-side Differential Privacy noise effect: {add_dp_noise}")
if add_dp_noise:
    print(f"  Simulated DP Noise Standard Deviation: {dp_noise_std}")
    print("  Note: Actual noise application would occur during aggregation in Step 6.")

# --- Data Structures for Metrics ---
# Initialize lists to store performance and simulation logs over rounds
global_train_loss = []
global_train_accuracy = []
global_test_accuracy = [] # Track test accuracy on the centralized test set
round_duration = []       # Log the time taken for each federated round
participating_clients_log = [] # Log the IDs of clients selected in each round

print("\nSimulation Configuration complete.")
```

```python
#@title Step 6: Training Loop Execution (Remove unused next() arguments)

print("\n--- Step 6: Training Loop Execution ---")

# --- Check if algorithms were built successfully in Step 4 ---
if federated_algorithm is None:
    print("Federated training algorithm was not built successfully in Step 4. Skipping training loop.")
elif num_total_clients == 0:
     print(f"num_total_clients is {num_total_clients}. Skipping training loop.")
     print("Ensure clients were successfully created during partitioning in Step 2.")
elif clients_per_round == 0:
     print(f"clients_per_round is {clients_per_round}. Skipping training loop.")
     print("Ensure clients_per_round is set to a value greater than 0.")
else:
    # --- Setup Evaluation Function (already done in Step 4, check here if needed) ---
    if evaluation_process is None:
        print("Federated evaluation process was not built successfully in Step 4. Training will proceed, but test evaluation will be skipped.")
    elif test_dataset is None: # Also check if test_dataset was loaded
         print("Test dataset was not loaded successfully. Test evaluation will be skipped.")
         evaluation_process = None # Ensure evaluation is skipped if test data is missing


    # Initialize the federated training state
    state = federated_algorithm.initialize()

    print("Initial state created.")

    # --- Start the training loop ---
    print("\nStarting federated training...")

    for round_num in range(1, num_rounds + 1):
        start_time = time.time()

        # --- Simulate Partial Participation ---
        if num_total_clients < clients_per_round:
             print(f"\nNot enough clients ({num_total_clients}) to select {clients_per_round} for round {round_num}. Stopping training.")
             break

        selected_client_ids = random.sample(train_client_ids, clients_per_round)
        participating_clients_log.append(selected_client_ids)

        print(f"\n--- Round {round_num}/{num_rounds} ---")
        print(f"  Selected clients ({len(selected_client_ids)}): {', '.join([cid for cid in selected_client_ids])}")

        # --- Prepare Client Data and Simulate Heterogeneity/Adversary ---
        client_data_list = []
        # In TFF 0.80.0 build_weighted_fed_avg, client weights and epochs
        # are typically determined by the dataset structure passed to .next()
        # or configured during the build phase. Remove explicit args here.

        for client_id in selected_client_ids:
            try:
                client_idx = train_client_ids.index(client_id)
                raw_client_dataset = train_client_datasets[client_idx]
            except ValueError:
                 print(f"Error: Could not find dataset for client ID {client_id}. Skipping client for this round.")
                 continue

            # We need to modify the client dataset *before* passing it to TFF
            # to include the correct number of local epochs (heterogeneity)
            # and potentially a weight component (though num_examples is default)

            # --- Simulate Heterogeneity: Apply local epochs to the dataset ---
            # TFF 0.80.0's build_weighted_fed_avg often expects client_epochs to be
            # configured via `client_epochs_fn` during the *build* step, or
            # specified as a single number, or handled if dataset has an 'epochs' element.
            # Passing varying epochs directly to `next` using client_work_args
            # seems to be the source of the "Keyword arguments not used" error.
            # Let's remove client_work_args and see if it trains with default epochs (usually 1).
            # Simulating *varying* epochs might require a more advanced TFF computation or process.

            # --- Simulate Adversary: Apply label flipping if this is the adversary client ---
            # This applies to the dataset structure before TFF processes it, so it should work.
            processed_data = preprocess_har_features(raw_client_dataset)
            if adversary_client_id is not None and client_id == adversary_client_id:
                 print(f"  Applying label flip attack for adversary client {client_id}...")
                 processed_data = apply_har_label_flip(processed_data)

            # Add the preprocessed dataset to the list for TFF's next step.
            # The client weight (num_examples) is usually inferred from the dataset itself.
            client_data_list.append(processed_data)


            # Add a small sleep for the straggler to *conceptually* simulate it being slow.
            # Note: Without explicit client_work_args/epochs to next, this sleep
            # or the local_epochs_per_client value for the straggler won't
            # actually make the *TFF simulation round* take longer, but the print
            # confirms the straggler was selected.
            if straggler_client_id is not None and client_id == straggler_client_id:
                print(f"  Straggler client {client_id} is participating (assigned {local_epochs_per_client.get(client_id, min_local_epochs)} local epochs conceptually).")
                # time.sleep(1)


        if not client_data_list:
            print(f"  No valid client data prepared from selected clients for round {round_num}. Skipping round.")
            round_duration.append(time.time() - start_time)
            global_train_loss.append(np.nan)
            global_train_accuracy.append(np.nan)
            global_test_accuracy.append(np.nan)
            continue

        # --- Execute one round of Federated Averaging ---
        try:
            # CORRECTED: Remove client_weights and client_work_args
            result = federated_algorithm.next(state, client_data_list) # Pass only state and data

            state = result.state
            train_metrics = result.metrics.get('client_work', {}).get('train', {})

            current_train_loss = train_metrics.get('loss', np.nan)
            current_train_accuracy = train_metrics.get('categorical_accuracy', np.nan)

            global_train_loss.append(current_train_loss)
            global_train_accuracy.append(current_train_accuracy)

            print(f"  Round {round_num} training metrics - Loss: {current_train_loss:.4f}, Accuracy: {current_train_accuracy:.4f}")

            # --- Secure Aggregation Simulation (Conceptual / Simple DP Noise) ---
            if add_dp_noise:
                 pass # Acknowledge conceptual step here.

        except Exception as e:
            print(f"  Error during training round {round_num}: {e}")
            print("  Simulating client failure/dropout handling: Round potentially failed or skipped.")
            global_train_loss.append(np.nan)
            global_train_accuracy.append(np.nan)
            global_test_accuracy.append(np.nan)
            round_duration.append(time.time() - start_time)
            continue

        # --- Evaluate Global Model on Test Set ---
        if evaluation_process is not None and test_dataset is not None:
            try:
                model_weights = state.model
                # Pass test_dataset as a list
                test_metrics = evaluation_process(model_weights, [test_dataset])
                current_test_accuracy = test_metrics.get('categorical_accuracy', np.nan)
                global_test_accuracy.append(current_test_accuracy)
                print(f"  Round {round_num} test metrics - Accuracy: {current_test_accuracy:.4f}")

            except Exception as e:
                print(f"  Error during test evaluation in round {round_num}: {e}")
                global_test_accuracy.append(np.nan)
        else:
             global_test_accuracy.append(np.nan)


        # --- Log Round Duration ---
        end_time = time.time()
        round_duration.append(end_time - start_time)
        print(f"  Round {round_num} duration: {round_duration[-1]:.2f} seconds")


print("\nFederated training simulation finished.")
```
